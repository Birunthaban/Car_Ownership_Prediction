## change datatypes of columns to correct format

```python
def change_dtype(element):
    final=""
    for c in str(element):
        if c.isnumeric():
            final+=c
    if final!= "":
         return float(final)  
    return py.nan
        
     
```

```python
type(py.nan)
```

```python
data['Monthly Income']= data['Monthly Income'].apply(change_dtype)
```

```python
data['Number of Children']=data['Number of Children'].apply(change_dtype)
```

```python
data['Years of Employment']=data['Years of Employment'].apply(change_dtype)
```
